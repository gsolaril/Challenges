## <a href = "https://www.hackerrank.com/challenges/grading/problem">Student grading</a>

```
function kangaroo(x1, v1, x2, v2) {
    var max_num = 1e4;
    console.assert((x1 < max_num) && (v1 < max_num)
                && (x2 < max_num) && (v2 < max_num))
    for (var n = 0; n <= max_num; n++) {
        x1 = x1 + v1; x2 = x2 + v2;
        if (x1 == x2) { return "YES"; } }
    return "NO"; }

x1 = 0; v1 = 3; x2 = 4; v2 = 2;
console.log("Input 1:", "x =", x1, "v =", v1);
console.log("Input 2:", "x =", x2, "v =", v2);
console.log("Output:", kangaroo(x1, v1, x2, v2));
```

## <a href = "https://leetcode.com/problems/single-number/">Only once</a>

```
function only_once_1(a) {
    arrset = Array.from(new Set(a));
    var once = [];
    for (var n = 0; n < arrset.length; n++) {
        var count = 0;
        count = a.filter(x => x == arrset[n]).length;
        if (count == 1) { once.push(arrset[n]) } }
    return once }

a = [1, 2, 3, 2, 3, 6]
console.log("Input:", a)
console.log("Output:", only_once(a))
```

```
function only_once_2(a) {
    a = a.sort((x1, x2) => x1 - x2);
    var once = [];
    for (var n = 1; n < a.length - 1; n++) {
        var alone_L = (a[n - 1] != a[n]);
        var alone_R = (a[n + 1] != a[n]);
        if (alone_L && alone_R) { once.push(a[n]); } }
    return once }

a = [1, 2, 3, 2, 3, 6]
console.log("Input:", a)
console.log("Output:", only_once(a))
```
