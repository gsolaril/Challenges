## <a href = "https://www.hackerrank.com/challenges/grading/problem">Student grading</a>

```
def gradingStudents(grades):
    return [(grade > 37 and grade % 5 > 2)*(5 - grade % 5) \
           + grade for grade in grades]

x = [4, 73, 67, 38, 33]
print("Input:", x)
print("Output:", gradingStudents(x))
```

## <a href = "https://www.hackerrank.com/challenges/apple-and-orange/problem">Apple & orange</a>

```
def countApplesAndOranges(s, t, a, b, apples, oranges):
    in_house = lambda fruit, tree: s <= tree + fruit <= t
    return (sum([in_house(apple, a) for apple in apples]),
            sum([in_house(orange, b) for orange in oranges]))

a, s, t, b = 5, 7, 11, 15
A, B = [-2, 2, 1], [5, -6]
print("Input:", A, ";", B)
print("Output:", countApplesAndOranges(s, t, a, b, A, B))
```

## <a href = "https://www.hackerrank.com/challenges/minimum-distances/problem">Minimum distances</a>

```
def minimumDistances(a):
    min_dist = float("inf")
    for id1 in range(len(a)):
        for id2 in range(id1 + 1, len(a)):
            if (a[id1] == a[id2]):
                min_dist = min(min_dist, id2 - id1)
                break
    if (min_dist == float("inf")): min_dist = -1
    return min_dist

a = [3, 2, 1, 2, 3]
print("Input:", a)
print("Output:", minimumDistances(a))
```
