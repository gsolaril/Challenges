## <a href = "https://www.hackerrank.com/challenges/grading/problem">Student grading</a>

```
function gradingStudents(grades) {
    var grades2 = [];
    for (var n = 0; n < grades.length; n++) {
        bonus = 5 - grades[n] % 5;
        cond = (bonus < 3) && (grades[n] > 37);
        grades2[n] = grades[n] + cond * bonus; }
    return grades2 }

x = [4, 73, 67, 38, 33]
console.log("Input:", x)
console.log("Output:", gradingStudents(x))
```

## <a href = "https://www.hackerrank.com/challenges/apple-and-orange/problem">Apple & orange</a>

```
function countApplesAndOranges(s, t, a, o, apples, oranges) {
    var n_a = 0; var n_o = 0;
    for (var n = 0; n < apples.length; n++) {
        n_a += (s <= a + apples[n]) && (a + apples[n] <= t) }
    for (var n = 0; n < oranges.length; n++) {
        n_o += (s <= o + oranges[n]) && (o + oranges[n] <= t) }
    return [n_a, n_o] }

a = 5; s = 7; t = 11; o = 15;
A = [-2, 2, 1] ; B = [5, -6]
console.log("Input:", A, ";", B)
console.log("Output:", countApplesAndOranges(s, t, a, o, A, B))
```

## <a href = "https://www.hackerrank.com/challenges/minimum-distances/problem">Minimum distances</a>

```
function minimumDistances(a) {
    var min_dist = Infinity
    for (var id1 = 0; id1 < a.length; id1++) {
        for (var id2 = id1 + 1; id2 < a.length; id2++) {
            if (a[id1] == a[id2]) {
                min_dist = Math.min(min_dist, id2 - id1);
                break; } } }
    if (min_dist == Infinity) { min_dist = -1; }
    return min_dist}

a = [3, 2, 1, 2, 3]
console.log("Input:", a)
console.log("Output:", minimumDistances(a))
```
